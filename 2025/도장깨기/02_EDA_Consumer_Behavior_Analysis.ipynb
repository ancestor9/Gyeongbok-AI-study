```python
# prompt: df의 컬럼명을 해당되는 nested_dict의 key를 찾아 해당 key의 값을 mapping하여 df의 값을 변경하라

import pandas as pd

def map_column_values(df, nested_dict):
    """
    Maps column values in a DataFrame based on a nested dictionary.

    Args:
        df: The input DataFrame.
        nested_dict: The nested dictionary containing the mapping.

    Returns:
        The DataFrame with mapped column values.
    """

    # Create a copy of the DataFrame to avoid modifying the original
    mapped_df = df.copy()

    for column in df.columns:
        if column in nested_dict:
            # Create a mapping dictionary for the current column
            mapping = nested_dict[column]

            # Use the mapping dictionary to update the column values
            mapped_df[column] = df[column].map(mapping)

    return mapped_df

# Example usage (assuming df and nested_dict are defined)
# Replace 'your_column_name' with the actual column name you want to map
mapped_df = map_column_values(df, nested_dict)
mapped_df
```
